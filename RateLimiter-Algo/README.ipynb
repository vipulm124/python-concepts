
# Rate Limited Algorithms


## Token Bucket:

#### Parameters:
1. Bucket size: the maximum number of tokens allowed in the bucket.

2. Refill rate: number of tokens put into the bucket every second.

Eg. Bucket Size = 4
    Refill Rate = 3

This means, every second, the bucket is checked for tokens, if there are already 4 tokens (Bucker Size), the no new tokens are added.

If there are 2 tokens, then 2 more are added to make count = 4

If there is 1 token, then 3 are added to make count = 4

So, at any point, the current bucket size cannot be > Bucket Size (here, 4).
And even if the bucket is empty, it cannot add more then 3 token (refill rate) per sec
