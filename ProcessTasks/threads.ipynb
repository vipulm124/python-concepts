```python
import time
import threading
from threading import Thread
import random
```

## Single Thread

```python
def worker():
    print("Worker thread started")
    time.sleep(12)
    print("Worker thread finished")
```

```python
t1 = Thread(target=worker)
```

```python
t1.name
```

```python
print(t1.ident) # this is None as thread is not yet started, id is generated once thread is started
```

```python
t1.start()
print(f"id of thread: {t1.ident}")
```

```python
# t1.join()
## if we uncomment the above line, the main thread will wait for t1 to complete before moving forward. So print('hi there!') will not be printed until t1 is done.
```

## Multi Thread

```python
def worker_2():
    print("Worker thread started")
    value = random.random() * 10
    print(f"Sleeping for {value} seconds")
    print(f"Current Thread: {threading.current_thread().name}")
    time.sleep(value)
    print("Worker thread finished")
```

```python
threads = [Thread(target=worker_2) for _ in range(5)]
```

```python
[t.start() for t in threads]
```

## Inherting thread

```python
class MyThread(Thread):
    def __init__(self, name=None):
        super().__init__(name=name)
    
    def run(self):
        identity= threading.get_ident()
        print(f"Thread here. Name: {self.name}. Id: {identity}")
        time.sleep(5)
        print("Existing thread now")
```

```python
t = MyThread("Vipul")
```

```python
t.start()
```

## Sharing data between threads in a same process

```python
SECONDS_TO_SLEEP = 15
```

```python
def thread_with_data():
    print("Starting thread")
    print(f"Sleeping now for {SECONDS_TO_SLEEP} secs")
    time.sleep(SECONDS_TO_SLEEP)
    print(f"New value of SECONDS_TO_SLEEP: {SECONDS_TO_SLEEP}")
    print("Terminating thread")
```

```python
threads  = [Thread(target=thread_with_data) for _ in range(10)]
```

```python
[t.start() for t in threads]
```
